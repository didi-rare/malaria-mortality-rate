```python
years = dict({
    'Year 2010': [year2010['Numeric'].sum(), americas2010['Numeric'].sum(), east2010['Numeric'].sum(),
                 euro2010['Numeric'].sum(), sea2010['Numeric'].sum(), west2010['Numeric'].sum()],
    'Year 2011': [africa2011['Numeric'].sum(), americas2011['Numeric'].sum(), east2011['Numeric'].sum(),
                 euro2011['Numeric'].sum(), sea2011['Numeric'].sum(), west2011['Numeric'].sum()],
    'Year 2012': [africa2012['Numeric'].sum(), americas2012['Numeric'].sum(), east2012['Numeric'].sum(),
                 euro2012['Numeric'].sum(), sea2012['Numeric'].sum(), west2012['Numeric'].sum()],
    'Year 2013': [africa2013['Numeric'].sum(), americas2013['Numeric'].sum(), east2013['Numeric'].sum(),
                 euro2013['Numeric'].sum(), sea2013['Numeric'].sum(), west2013['Numeric'].sum()],
    'Year 2014': [africa2014['Numeric'].sum(), americas2014['Numeric'].sum(), east2014['Numeric'].sum(),
                 euro2014['Numeric'].sum(), sea2014['Numeric'].sum(), west2014['Numeric'].sum()],
})
extract_data = pd.DataFrame(years, index=['Africa', 'Americas', 'Eastern Mediterranean', 'Europe', 
                                          'South-East Asia', 'Western Pacific'])
extract_data
```
